```python
#중복 제거하기 방법 2

l = []
while True:
    l.append(random.randint(1,45))
    l = set(l)
    l = list(l)
    if len(l) == 6:
        break
print(l)
```

```python
class Asia:
    def __init__(self, name):
        self.name = name

    def show(self):
        return '해당 국가는 아시아에 있습니다.'

class Korea(Asia):
    def __init__(self, name, capital, population):
        Asia.__init__(self,name)                        #이름은 부모로부터 상속 받은 함수
        self.capital = capital
        self.population = population

    def showname(self):
        return f'국가이름은 {self.name}'
```

```python
a=Korea('대한민국','오천만','서울')
a.showname()
```

```python
#예외처리
# 2~9 구구단
#2보다 작으면 2단 출력, 9보다 크면 9단출력, 정수가 아니면 예외문장 출력

num = input('정수를 입력해주세요')

try:
    if(num.isdigit()):
        num = int(num)
    
    if(num > 9):
        num = 9
    if(num < 2):
        num = 2
    else:
        num = num 
    
    for i in range(1,10):
        print('{} x {} =  {}'.format(num, i , num*i))

except Exception as msg:
    print("예외발생 : ", msg)
        
```
